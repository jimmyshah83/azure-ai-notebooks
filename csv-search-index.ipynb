```python
import pandas as pd

df = pd.read_csv('data/candidate_experience/Open_Critical_Care_Nursing.csv', header=0)
```

```python
import requests
import os

url = df['Job Description'].tolist()

for i in url:
    filename = i.split('/')[-1]
    response = requests.get(i)
    folder_path = "data/candidate_experience/pdfs"
    with open(os.path.join(folder_path, filename), "wb") as file:
        file.write(response.content)
```

```python
import os
import dotenv

dotenv.load_dotenv()

search_baseUrl = os.getenv("SEARCH_BASE_URL")
api_key = os.getenv("SEARCH_API_KEY")
```

```python
import PyPDF2

def convert_pdf_to_text(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in reader.pages:
            text += page.extract_text()
    return text
```

```python
for index, row in df.iterrows():
    filename = row['Job Description'].split('/')[-1]
    pdf_path = 'data/candidate_experience/pdfs/' + filename
    df.at[index, 'Job Description'] = convert_pdf_to_text(pdf_path)
```

```python
df.to_csv('data/candidate_experience/Open_Critical_Care_Nursing_updated.csv', index=False)
```

```python
selected_columns = df[['Job Title', 'Salary Grade', 'Location', 'Additional Information For Applicants', 'Job Description']]
selected_columns.rename(columns={'Job Title': 'title', 'Salary Grade': 'pay', 'Location': 'location', 'Additional Information For Applicants': 'add_info', 'Job Description': 'jd'}, inplace=True)
```

```python
selected_columns.fillna('', inplace=True)
selected_columns
```

```python
from openai import AzureOpenAI

client = AzureOpenAI(
  api_key = os.getenv("AZURE_OPENAI_API_KEY"),  
  api_version = "2024-02-01",
  azure_endpoint =os.getenv("AZURE_OPENAI_ENDPOINT") 
)
```

```python
def create_embedding(text):
    return client.embeddings.create(
        input = text,
        model= os.getenv("EMBEDDING_DEPLOYMENT_NAME")
    ).data[0].embedding
```

```python
import uuid

def upload_data(row):
 
    base_url=f"{search_baseUrl}/indexes/{os.getenv("SEARCH_INDEX_NAME")}/docs/index?api-version=2023-11-01"
    headers = {"Content-Type": "application/json", "api-key": api_key}
    body = {
    "value": [
        {
            "@search.action": "upload",
            "id": str(uuid.uuid4()),
            "title": row['title'],
            "add_info": row['add_info'],
            "location": row['location'],
            "salary": row['pay'],
            "chunk": row['jd'],
            "chunkVector": create_embedding(row['jd'])
        }
    ]
    }
    return requests.post(url = base_url, headers = headers, json = body)
```
